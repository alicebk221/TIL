```python
import numpy as np
import matplotlib.pyplot as plt

# 함수() 안에서 shift+tab 누르면 함수에 대한 정보 나옴
# 함수. + tab 누르면 쓸 수 있는 메소드? 알 수 있음
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
plt.plot(x, y)
plt.title("y = x ** 2")   # plot 이름
plt.xlabel("int var")     # x 라벨
plt.ylabel("dep var")     # y 라벨
```

```python
plt.subplot(1, 2, 1) # 1행 2열의 그래프를 그리고 그 중에 첫 번째 그래프 그리기
plt.plot(x, y)       
plt.title("y = x ** 2")
plt.subplot(1, 2, 2)
plt.plot(y, x)
plt.title("y = x**(0.5)")
```

```python
plt. subplot(2, 2, 1)
plt.plot(x, y)       
plt.title("1")
plt.subplot(2, 2, 2)
plt.plot(y, x)
plt.title("2")
plt.subplot(2, 2, 3)
plt.plot(y, x*2)
plt.title("3")
plt. subplot(2, 2, 4)
plt.plot(x, y)       
plt.title("4")
```

## Object Oriented Method (객체 지향)
1. Figure 생성
2. 해당 Figure에 axes를 Add

```python
fig = plt.figure()  # Figure 생성
type(fig)
# axes를 생성하고 캔버스 위치 설정 : 왼쪽, 오른쪽(캔버스 내에 어디에 그래프를 그릴 것인지), 가로, 세로 (캔버스의 크기)
axes = fig.add_axes([0, 0, 1, 1])
axes       # 필요할 때 이 axes를 불러오면 캔버스처럼 활용 가능함 -> axes마다 고유 번호가 부여됨
axes.plot(x, y)
axes.set_title("y = x ** 2")
axes.set_xlabel("int var")
axes.set_ylabel("dep var")
```

```python
# 그래프 겹쳐서 보여주기
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.1, 0.5, 0.5, 0.4])
axes1.plot(x, y)
axes2.plot(y, x)
axes2.set_title("y = x^(0.5)")
```

```python
fig, axes = plt.subplots(1, 2)    # 결과물이 tuple형태로 들어가기 때문에 가능
# fig에는 전체 캔버스, axes에는 array가 저장된다
```

```python
axes[0].plot(x, y)
axes[1].plot(y, x)
fig
```

```python
# 범례 : legend
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label = "y = x ** 2")
ax.plot(x, x**3, label = "y = x ** 3")
ax.legend()   # 범례 표시, (loc = 0~10)으로 직접 위치 지정 가능
# ax.legend(loc = [0.1, 0.4]), ax.legend(loc = [1.5, 0.5]) 등의 위치 지정도 모두 가능
```

```python
plt.figure()
plt.plot(x, y, 'cs--')   # b : 색 지정, x : 마커 모양 지정(s, o, p 등), - : 선의 유무(-- : 점선)
```

```python
plt.figure()
plt.plot(x, y, color = 'red', marker = '*', ls = '-.', lw = 0.5) # lw : 라인 굵기, ms : 마커 크기
```

```python
# np.arange(n) : n까지의 값을 특정 array에 넣는 것
for i in np.arange(50):
    print(i)
```

```python
np.random.randn(100)   # normal distribution 내에서 랜덤값 출력
```

```python
plt.scatter(np.arange(100), np.random.randn(100))  # x : 0~100까지, y : 정규분포 내에서 랜덤 숫자
```

```python
# randn이 정말 정규분포 특성을 가지고 있는지 판단하기 위해 히스토그램 그려보기
# hist(x) : x값을 넣었을 때 히스토그램을 그려준다
# bins : 막대 하나의 간격(?) 숫자가 클 수록 그래프가 촘촘해짐
plt.hist(np.random.randn(5000), bins=100) 
```

```python
x = np.arange(100)
y = x
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.1, 0.5, 0.4, 0.4])
ax2.set_title("zoom")
ax1.plot(x, y*2)
ax2.plot(x, y*2)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
# 범위 한정
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)
```
