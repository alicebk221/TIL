```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

# tips 데이터 실습

```python
sns.set()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape    # 244개의 행, 7개의 열
```

```python
sns.relplot(x = "total_bill", y = "tip", data=tips)
```

```python
sns.relplot(x = "total_bill", y = "tip", data=tips, col = "time")
```

```python
sns.relplot(x = "total_bill", y = "tip", data=tips, col = "time", hue="smoker")
```

```python
sns.relplot(x = "total_bill", y = "tip", data=tips, col = "time", hue="smoker", style="smoker", size="size")
```

# 실습 2

```python
time = np.arange(500)
value = np.random.randn(500).cumsum()
```

```python
df = pd.DataFrame(dict(time = time, value = value))
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df)  # kind를 지정하지 않으면 default는 scatter
```

# fmri 실습

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
# ci = None : 신뢰구간 표시 제외
# estimator = None : 추정치를 반영하지 않고 데이터 그대로 뿌리는 것
sns.relplot(x = "timepoint", y = "signal", kind = "line", hue = "event", style = "event", ci = None,
            data = fmri)
```

# tips 실습
sns.catplot()

```python
# kind = "swarm" : 데이터 밀도가 높은 구간을 펼쳐서 보여줌
sns.catplot(x = "day", y = "total_bill", kind = "swarm", data = tips) # strip plot
```

```python
sns.catplot(x = "day", y = "total_bill", kind = "swarm", hue = "sex", data = tips) # swarm plot
```

```python
sns.catplot(x = "day", y = "total_bill", kind = "swarm", hue = "time", data = tips)
```

```python
sns.catplot(x = "day", y = "total_bill", kind = "swarm", hue = "sex", col = "time", data = tips)
```

```python
sns.catplot(x = "day", y = "total_bill", kind = "box", data = tips)  # box plot
```

```python
sns.catplot(x = "day", y = "total_bill", kind = "violin", data = tips) # violin plot
sns.catplot(x = "day", y = "total_bill", hue = "sex", kind = "violin", split = True, data = tips) 
```

# diamonds 실습

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
sns.catplot(x = "color", y = "price", kind = "boxen", data = diamonds)
```

```python
sns.distplot(np.random.normal(size = 200), bins = 50)
```

```python
# kde(kernel density estimator )
sns.distplot(np.random.normal(size = 200), bins = 50, kde = False)
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips)
```

```python
sns.jointplot(x = "total_bill", y = "tip", kind = "reg", data = tips) # regression 추기
```

```python
# 모든 변수를 하나의 figure 안에서 보여줌, 대각선축을 기준으로 대칭인 그래프
iris = sns.load_dataset('iris')
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips) # regression 그래프
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'time', data = tips)
```

```python
sns.lmplot(x = 'total_bill', y = 'tip', row = 'smoker', col = 'time', data = tips)
```
