```python
import pandas as pd
import numpy as np
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

px : 정형화된 데이터만 가능  
go : 비정형화된 데이터도 가능

```python
iris = px.data.iris()
iris.head()
```

```python
px.scatter(iris, x = 'sepal_length', y = 'sepal_width', color = 'species', size = 'petal_length')
```

```python
data = px.data.gapminder()
data.head()
```

```python
data['country'].unique()
```

```python
canada = data[data['country'] == 'Canada']
px.bar(canada, x = 'year', y = 'pop', color = 'lifeExp')
```

```python
px.line(canada, x = 'year', y = 'lifeExp')
```

```python
data = data.query("continent == 'Asia'")
px.line(data, x = 'year', y = 'lifeExp', color = 'country')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/stockdata2.csv')
```

```python
msft = df[df['stock'] == 'MSFT']
appl = df[df['stock'] == 'AAPL']
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = msft['Date'], y = msft['value'], mode = 'lines', name = 'MSFT'))
fig.add_trace(go.Scatter(x = appl['Date'], y = appl['value'], mode = 'lines', name = 'AAPL'))
```
