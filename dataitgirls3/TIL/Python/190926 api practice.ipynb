```python
import os
key = os.environ['WEATHER_API_KEY']
```

```python
from urllib.parse import unquote
API_KEY = unquote(key)
```

```python
import requests

requests.__version__
```

```python
import pandas as pd
```

```python
url = 'http://openapi.airkorea.or.kr/openapi/services/rest/ArpltnInforInqireSvc/getCtprvnMesureSidoLIst'

payload = {
    'serviceKey' : API_KEY,
    'numOfRows' : 1000,
    'sidoName' : '경기',
    'searchCondition' : 'DAILY'
}

response = requests.get(url, params=payload)
```

```python
import xml.etree.ElementTree as ET

root = ET.fromstring(response.text)

for item in root.iter('item'):
    city = item.find('cityName').text
    pm10 = item.find('pm10Value').text
    pm25 = item.find('pm25Value').text
    print(f'{city} | {pm10} | {pm25}')
```

```python
item = next(root.iter('item'))
```

```python
list(item)
```

```python
columns = []

for child in list(item):
    columns.append(child.tag)
    
print(columns)
```

```python
rows = []

for item in root.iter('item'):
    row = {}
    for child in list(item):
        row[child.tag] = child.text
    rows.append(row)
    
print('갯수', len(rows))
print()
print(rows[0])
```

```python
df = pd.DataFrame(rows, columns=columns)

df
```

```python
data = df['pm10Value'].value_counts()
pd.DataFrame(data).sort_index().head(10)
```
